```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#define image size and batch size
IMG_SIZE=224
BATCH_SIZE=32
```

```python
#Define data generators for train,validation and test sets
train_datagen=ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

train_generator=train_datagen.flow_from_directory(r'C:\Users\JyothiSwaroop\Downloads\melanoma_cancer_dataset\train',
    target_size=(IMG_SIZE,IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='training'
)
val_generator=train_datagen.flow_from_directory(r'C:\Users\JyothiSwaroop\Downloads\melanoma_cancer_dataset\train',
    target_size=(IMG_SIZE,IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
subset='validation')
val_test_generator=train_datagen.flow_from_directory(r'C:\Users\JyothiSwaroop\Downloads\melanoma_cancer_dataset\test',
    target_size=(IMG_SIZE,IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
subset='validation')
```

```python
#define the model
model = keras.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',input_shape=(IMG_SIZE,IMG_SIZE,3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(128,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128,activation='relu'),
    layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history=model.fit(train_generator,validation_data=val_generator,epochs=1)
```

```python
model.save("skin.h5","label.txt")
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np

#load the saved model
model=load_model(r'c:\Users\JyothiSwaroop\skin.h5')

#load and preprocess the test image
test_image_path=r'C:\Users\JyothiSwaroop\Downloads\melanoma_cancer_dataset\test\predict\melanoma_10597.jpg'
img=image.load_img(test_image_path,target_size=(224,224))
img_array=image.img_to_array(img)
img_array=np.expand_dims(img_array ,axis=0)  #batch dimension
img_array /=255.  #normalise the pixel values
#make predictions
predication=model.predict(img_array)
##print the predication
if predication < 0.5:
  print("predication: no skincancer present !(probility:",predication[0][0],")")
else:
    print("predication :skincancer present! (probility:",predication[0][0],")")
```
