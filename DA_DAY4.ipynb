```python
url1="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
```

```python
req1=requests.get(url1)
```

```python
s2=BeautifulSoup(req1.text,'html')
print(s2)
```

```python
t=s2.find('table')
print(t)
```

```python
heading=t.find_all("th")
print(heading)
```

```python
rows=t.find_all("tr")
print(rows)
```

```python
column=t.find_all("td")
print(column)
```

```python
head = [i.text for i in heading]
print(head)
```

```python
mydata=pd.DataFrame(columns=head)
print(mydata)
```

```python
for i in rows[1:9]:
  rd =i.find_all("td")
  er=[i.text for i in rd]
  print(er)
```
